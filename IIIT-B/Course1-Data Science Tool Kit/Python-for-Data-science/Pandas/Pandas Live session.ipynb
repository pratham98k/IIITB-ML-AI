```python
#Group By
#it is used to see aggregated result on the dataframe
for i in range(1,13):
    print(i, round(df[df["mnth"]==i]["temp"].mean(),2))
```

```python
df.groupby("mnth")["temp"].mean()
#mnth: Group column
#temp: Aggregating Column
#mean: Aggregating function
```

```python
df.groupby("mnth")[["temp","hum"]].mean()
```

```python
df.groupby("mnth")[["temp","hum"]].median()
```

```python
d={"hum":["max","min","mean"],"temp":["median","std"],"windspeed":["count"]}
df.groupby("mnth").agg(d)
```
