```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", 300)
pd.set_option("display.max_rows", 500)
```

```python
df = pd.read_csv("popularity.csv")
```

```python
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
df["shares"]
```

```python
sns.boxplot(df["shares"])
```

```python
# Box plot
plt.boxplot(df['shares'])
plt.show()
```

```python
plt.scatter(df.index, df['shares'])
plt.show()
```

```python
# Calculate Q1 (25th percentile) and Q3 (75th percentile)
Q1 = df['shares'].quantile(0.25)
Q3 = df['shares'].quantile(0.75)
IQR = Q3 - Q1

# Define outliers
outliers = df[(df['shares'] < (Q1 - 1.5 * IQR)) | (df['shares'] > (Q3 + 1.5 * IQR))]

print(outliers)
```
