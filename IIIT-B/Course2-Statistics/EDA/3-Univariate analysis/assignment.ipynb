```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("tendulkar_ODI.csv")
```

```python
df
```

```python
df.info()
```

```python
df["Runs"]=df["Runs"].str.replace("*", "")
```

```python
df["Runs"] = df["Runs"].astype("int32")
```

```python
df["Runs"] = pd.to_numeric(df["Runs"], errors='coerce')
```

```python
df["Runs"].fillna(0, inplace=True)
```

```python
df["Runs"] = df["Runs"].astype("int32")
```

```python
run_frequency= df["Runs"].value_counts().sort_index()
```

```python
sns.barplot(x=run_frequency, y=df["Runs"])
```

```python
# Step 1: Create a frequency distribution of the 'Runs' column
runs_frequency = df["Runs"].value_counts().sort_index()

# Step 2: Plot the bar chart
plt.figure(figsize=(10, 6))
runs_frequency.plot(kind='bar', color='skyblue')

# Step 3: Add labels and title
plt.xlabel('Runs Scored')
plt.ylabel('Frequency')
plt.title('Frequency of Runs Scored')
plt.xticks(rotation=0)

# Show the plot
plt.show()

# Step 4: Identify the bucket with the most frequent score
most_frequent_score = runs_frequency.idxmax()
most_frequent_count = runs_frequency.max()
```
